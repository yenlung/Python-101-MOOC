```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 6)
y = np.random.rand(6) * 2 - 1
```

```python
plt.scatter(x, y, s=60, c='r');
```

```python
X = np.linspace(0, 1, 2000)
```

```python
def my_fcn(n=1):
    Y = np.sin(n*X)
    plt.scatter(x, y, s=60, c='r')
    plt.plot(X, Y)
```

```python
my_fcn()
```

```python
from ipywidgets import interact
```

```python
interact(my_fcn, n=(1, 500));
```
