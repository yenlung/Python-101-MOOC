## 10. 支援 `LaTeX` 語法

`LaTeX` 是很標準的、可以打出漂亮數學符號的工具。

只要把數學式子「前後加錢」就可以。前後各加兩個 `$` 可以變成「展示模式」, 把式子置中強調。

    我們有一個函數 $f(x) = x^2$, 來積分一下:

    $$\int_a^b f(x) \, dx$$

我們有一個函數 $f(x) = x^2$, 來積分一下:

$$\int_a^b f(x) \, dx$$

#### 【注意】LaTeX 的發音

`LaTeX` 結尾沒有 `s` 的音, 請參考 [TeX FAQ](https://texfaq.org/FAQ-latexpronounce)。

#### 【注意】函數名稱要正體

比如說 `$sin(x)$`, 會顯示成 $sin(x)$; 正確應該要打成 `$\sin(x)$`, 也就是正體顯示函數名稱: $\sin(x)$。
