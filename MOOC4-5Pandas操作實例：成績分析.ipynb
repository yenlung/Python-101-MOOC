```python
df = pd.read_csv("https://bitly.com/gradescsv")
```

```python
df["三科總級分"] = df[["國文", "英文", "數學"]].sum(axis=1)
```

```python
df.head()
```

```python
df["甄選總級分"] = df.國文 + df.英文 * 1.5 + df.數學 * 2
```

```python
df.head()
```

```python
df.sort_values("甄選總級分", ascending=False)
```

```python
df = df.sort_values("甄選總級分", ascending=False)
```
