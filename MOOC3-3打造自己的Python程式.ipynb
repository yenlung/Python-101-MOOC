## 2. 打造一個拍拍機器人

```python
def chat_completion(prompt):
    response = "拍拍"
    return response
```

```python
bot_avatar = "(*¯︶¯*)"
```

```python
print(f"{bot_avatar}: 你好，我是拍拍，今天想聊什麼呢?")
print()
```

```python
while True:
    prompt = input("> ")
    if "bye" in prompt:
        print(f"{bot_avatar}: bye bye")
        print()
        break
    reply = chat_completion(prompt)
    print(f"{bot_avatar}: {reply}")
    print()
```
