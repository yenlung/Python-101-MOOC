## 4. Ê•µÂ∫ßÊ®ôÊèõÊàêÂèÉÊï∏Âºè

\begin{cases}
x = r \cdot \cos(t) \\
y = r \cdot \sin(t)
\end{cases}

ÂÖ∂‰∏≠ $t$ ÊòØËßíÂ∫¶, ‰∏ÄËà¨ÊòØ $0$ Âà∞ $2 \pi$„ÄÇ

ÊØîÂ¶ÇË™™Ôºå‰∏ÄÂÄãÂçäÂæëÊòØ 3 ÁöÑÂúìÊòØ $r=3$ÔºåÊèõÊàêÊàëÂÄëÁøíÊÖ£ÁöÑÂ∫ßÊ®ôÁ≥ªÁµ±Â∞±ÊòØ:

$t$ Âèñ $0$ Âà∞ $2\pi$ ÁöÑËã•Âπ≤ÈªûÔºå
\begin{cases}
x = 3 \cdot \cos(t)\\
y = 3 \cdot \sin(t)
\end{cases}


```python
pi = np.pi
```

```python
t = np.linspace(0, 2*pi, 400)
```

```python
r = 3
```

```python
x = r * np.cos(t)
y = r * np.sin(t)
```

```python
plt.plot(x,y)
```

```python
plt.axis("equal")
plt.plot(x,y)
```
