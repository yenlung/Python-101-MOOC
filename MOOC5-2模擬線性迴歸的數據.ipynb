```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 50)
y = 1.2*x + 0.8
```

```python
plt.scatter(x, y);
```

```python
y = 1.2*x + 0.8 + 0.4*np.random.randn(50)
plt.scatter(x, y)
plt.plot(x, 1.2*x + 0.8, 'r');
```

```python
plt.scatter(x, y);
```

```python
x
```

```python
x_train = x.reshape(-1, 1)
```

```python
x_train.shape
```

```python
y_train = y.reshape(-1, 1)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict([[1.3]])
```

```python
model.predict([[1.3], [2.71828]])
```

```python
x = x_train.ravel()
y = y_train.ravel()
y_predict = model.predict(x_train).ravel()

plt.scatter(x, y)
plt.plot(x, y_predict, 'r');
```

```python
r_squared = model.score(x_train, y_train)
print(f'R-squared: {r_squared}')
```
