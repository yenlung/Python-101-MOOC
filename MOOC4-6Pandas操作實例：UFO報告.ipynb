```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://bitly.com/uforeports")
```

```python
df.head()
```

```python
df_state = df.groupby("State").count()
```

```python
df_state.head()
```

```python
df_state.sort_values("Time", ascending=False, inplace=True)
```

```python
df_state.head()
```

```python
df_state[:10].Time.plot(kind="bar")
```
